# Project: Customer Retention & Revenue Impact Analysis

## Phase 1: Business Understanding

**Objective:**  
Identify customers likely to churn (stop purchasing) and estimate potential revenue at risk, so businesses can take preventive action.

**Why This Matters:**  
Retaining existing customers is cheaper than acquiring new ones. Predicting churn protects revenue and improves customer relationships.

**Key Business Questions:**
- Who is likely to churn?
- What behaviors indicate churn?
- How much revenue is at risk?
- Which customers should be targeted first?

**Success Metrics:**
- Churn rate
- Revenue at risk
- Model precision & recall

**Datasets Used:**
1. Retail Transaction Data
2. Customer Support Ticket Data
3. Telecom Customer Churn Data

**Tools Used:**
- Python (Pandas, Matplotlib)
- SQL (SQLite)
- Power BI (for dashboards)
